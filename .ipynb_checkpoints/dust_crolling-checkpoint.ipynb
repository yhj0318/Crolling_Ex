```python
html = requests.get('https://search.naver.com/search.naver?query=%EB%82%A0%EC%94%A8')
pprint(html.text)
```

```python
pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(html.text, 'html.parser')
pprint(soup)
```

```python
data1 = soup.find('ul', {'class':'today_chart_list'})
# 데이터를 저장할 변수를 생성하고 find함수를 사용한다.
# 찾고싶은 데이터를 감싸고 있는 큰 태그를 적고, 딕셔너리로 클래스에
# 해당하는 속성의 이름을 적어주면 된다.
# 같은 속성의 이름일 경우 첫 번째만 탐색이 된다. 두 번째 세 번째는 무시가 됨
pprint(data1)
```

```python
data2 = data1.findAll('li')
pprint(data2[0])
# findAll함수는 데이터를 쪼개는 역할을 할 수 있다. 배열 형태로 만들 수 있음
```

```python
find_dust = data2[0].find('span', {'class':'txt'})
pprint(find_dust.text)
# .text를 하면 span태그에 있는 텍스트만 뽑을 수 있음
```
