```python
from pprint import pprint
import requests
from bs4 import BeautifulSoup as bs
```

```python
html = requests.get('https://media.naver.com/press/025?sid=105')
pprint(html.text)
```

```python
soup = bs(html.text, 'html.parser')
```

```python
data1 = soup.find('div',{'class':'_tab_panel'})
pprint(data1)
```

```python
data2 = data1.findAll('span', {'class':'press_edit_news_title'})
pprint(data2)
```

```python
title_list=[]
```

```python
for t in data2:
    title_list.append(t.text)
# 텍스트만 추출하는 방법
# title_list = [t.text for t in data2] 리스트로 사용하면 간단하다
```

```python
pprint(title_list)
```

```python
find_AI = [s for s in title_list if "AI" in s]
```

```python
pprint(find_AI)
```

```python
pprint(find_AI.count("AI"))
```

```python
pprint(len(find_AI))
```
